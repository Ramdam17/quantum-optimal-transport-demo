## 2. Your first quantum circuit

Let's prove the quantum stack works from end to end. We'll build the smallest interesting circuit — a single qubit, a **Hadamard** gate, and a measurement — and run it on a local simulator. You do not need to understand the gate yet (that is what `01_QuantumFoundations` is for); right now we only want to watch it run and return numbers.

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

qc = QuantumCircuit(1, 1)
qc.h(0)             # put the qubit into an equal superposition
qc.measure(0, 0)    # read it out in the computational basis

simulator = AerSimulator()
result = simulator.run(transpile(qc, simulator), shots=2048, seed_simulator=42).result()
counts = result.get_counts()
print("measurement counts:", counts)

viz.plot_counts(counts)
plt.show()
```

**Read the figure.** The bars count how often each outcome (`0` or `1`) appeared across 2048 repetitions. A Hadamard gate turns a definite `0` into a fair coin, so you should see roughly 1024 of each — close to 50/50, with a little wobble from random sampling. Two bars of comparable height mean your simulator, your plotting, and the course style are all working together. That wobble has a name (shot noise) and we will meet it properly in module 01.
