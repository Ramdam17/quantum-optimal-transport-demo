# 10 — Entanglement

Build a two-qubit state that is perfectly definite as a whole, yet whose parts are maximally uncertain — something no classical joint distribution can do.

**Prerequisites:** `01/09_tensor_and_partial_trace`, `01/06_pure_vs_mixed`.

**What you'll be able to do**
- Recognise an entangled state as one that is not a product.
- Show a Bell state is pure, yet each qubit alone is maximally mixed.
- Read the entanglement entropy as the mixedness of a part.

In the last notebook, a product state's parts were perfectly well-defined. Now we break that. The Bell state is as definite as a state can be — purity $1$ — yet ask about either qubit alone and you get total uncertainty. That gap between whole and parts is **entanglement**, the headline resource of quantum information.

```python
import numpy as np
import matplotlib.pyplot as plt

from qot_course import viz
from qot_course.quantum.density import density_matrix, maximally_mixed, purity
from qot_course.quantum.composite import bell_state, partial_trace, entanglement_entropy

np.random.seed(42)
viz.use_course_style()
```

## 1. A state that is not a product

Some two-qubit states cannot be written as $|\psi_A\rangle \otimes |\psi_B\rangle$ at all. The famous example is the **Bell state**

$$ |\Phi^+\rangle = \tfrac{1}{\sqrt2}\big(|00\rangle + |11\rangle\big). $$

Read it as: the two qubits are guaranteed to *agree*, but *which* value they share is left undecided. There is no way to factor this into a state for A times a state for B. Let's build it and take its partial trace.

```python
rho_bell = density_matrix(bell_state())
reduced_A = partial_trace(rho_bell, keep=[0], dims=[2, 2])

print("global purity (Bell):", round(purity(rho_bell), 3), " -> pure")
print("reduced A == I/2? ", np.allclose(reduced_A, maximally_mixed(2)))
print("entanglement entropy:", round(entanglement_entropy(rho_bell, dims=[2, 2]), 3), "bit")

two_qubit_basis = ["|00>", "|01>", "|10>", "|11>"]
qubit_basis = ["|0>", "|1>"]
viz.plot_density_matrix(rho_bell, title="Bell state rho (global) — pure", basis_labels=two_qubit_basis)
viz.plot_density_matrix(reduced_A, title="Reduced state of qubit A — I/2 (maximally mixed)", basis_labels=qubit_basis)
plt.show()
```

**The punchline — read both figures together.** The global Bell state is **pure** (purity $1$; the off-diagonal corners hold the $|00\rangle$–$|11\rangle$ coherence). Yet each qubit *alone* is **maximally mixed** — $I/2$, a full bit of entropy. The whole is perfectly defined while its parts are maximally uncertain. No classical joint distribution can manage this: classically, if the whole is certain then every part is certain too. This is **entanglement** — the resource that makes quantum information genuinely different, and, later, the structure that gives quantum optimal transport something no diagonal-only view can see.
