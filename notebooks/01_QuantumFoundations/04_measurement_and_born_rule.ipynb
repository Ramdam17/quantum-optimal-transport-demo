You can write a qubit's amplitudes, but you can never read them off directly. Nature only lets you *measure*, and a measurement returns a plain $0$ or $1$, at random. The rule connecting the two is short, exact, and one of the most important sentences in the course.

```python
import numpy as np
import matplotlib.pyplot as plt

from qot_course import viz
from qot_course.colors import COLORS
from qot_course.quantum.states import KET_0, KET_PLUS, qubit_state, born_probabilities

np.random.seed(42)
viz.use_course_style()
```

## 1. The Born rule

When you measure a qubit in the computational basis, you get outcome $x \in \{0, 1\}$ with probability equal to the **squared size of its amplitude**:

$$ P(x) = |a_x|^2. $$

This is the **Born rule**. Two consequences are worth saying out loud. First, measurement is **random**: even a perfectly known state yields an unpredictable single outcome. Second, because we square the amplitude, the probabilities depend on its *size* — and, as we are about to see, not at all on its phase.

```python
psi = qubit_state(theta=2 * np.pi / 3, phi=np.pi / 4)
for name, state in [("|0>", KET_0), ("|+>", KET_PLUS), ("|psi>", psi)]:
    p = born_probabilities(state)
    print(f"{name:5s} P(0) = {p['0']:.3f}   P(1) = {p['1']:.3f}")
```

**Read the output.** $|0\rangle$ always gives $0$ ($P(0)=1$) — a classical state is certain. $|+\rangle$ is a perfect coin flip, $0.5/0.5$ — the equal superposition is maximally uncertain. The tilted $|\psi\rangle$ lands somewhere in between. Let's see all three as a picture.

```python
states = {"|0>": KET_0, "|+>": KET_PLUS, "|psi>": psi}
p0 = [born_probabilities(s)["0"] for s in states.values()]
p1 = [born_probabilities(s)["1"] for s in states.values()]

fig, ax = plt.subplots(figsize=(7.5, 4.2))
x = np.arange(len(states))
width = 0.38
ax.bar(x - width / 2, p0, width, color=COLORS["source"], label="P(0)")
ax.bar(x + width / 2, p1, width, color=COLORS["target"], label="P(1)")
ax.set_xticks(x)
ax.set_xticklabels(list(states))
ax.set_ylabel("probability")
ax.set_ylim(0, 1)
ax.set_title("Born-rule probabilities for a Z measurement", pad=12)
ax.legend()
plt.show()
```

**Read the figure.** $|0\rangle$ puts all its probability on outcome $0$. $|+\rangle$ splits $50/50$ — a fair coin. $|\psi\rangle$ leans to one side. For every state the two bars sum to $1$: you always get *some* outcome. This pair of numbers is the whole story of what a single $Z$ measurement can tell you about a state.

## 2. What the measurement cannot see

Here is the subtle part. Keep a state on the equator ($\theta = \pi/2$, an equal superposition) and vary only the phase $\phi$. Geometrically — from notebook 02 — this **spins the state around the equator**, landing on a genuinely different point of the Bloch sphere each time. Does the $Z$ measurement notice?

```python
for phi in [0.0, np.pi / 2, np.pi]:
    state = qubit_state(theta=np.pi / 2, phi=phi)   # equator; only the phase changes
    p = born_probabilities(state)
    print(f"phi = {phi:.2f}:  P(0) = {p['0']:.3f}   P(1) = {p['1']:.3f}")
```

**Read the output.** The phase $\phi$ sweeps the state a quarter- and then a half-turn around the equator, yet $P(0)$ and $P(1)$ never move from $0.5$. A computational-basis measurement is **blind to phase**. That information is real — it lives in the off-diagonal of the density matrix you will build in `01/05`, and *other* measurements can see it — but a plain $Z$ measurement throws it away. Remember this: it is exactly the gap that classical descriptions miss and that quantum optimal transport will eventually care about.
