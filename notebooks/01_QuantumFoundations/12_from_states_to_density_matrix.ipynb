```python
import numpy as np
import matplotlib.pyplot as plt

from qot_course import viz
from qot_course.quantum import density
from qot_course.quantum.states import KET_0, KET_PLUS

np.random.seed(42)
viz.use_course_style()
```

## 1. From a vector to an operator

A pure state $|\psi\rangle$ becomes a **density matrix** through the outer product

$$ \rho = |\psi\rangle\langle\psi|, $$

an operator (a matrix) rather than a vector. For one qubit it is a $2\times2$ complex matrix. Why promote a perfectly good vector? Because $\rho$ can *also* describe **mixtures** — classical uncertainty about which pure state we have — which no single vector can represent. We build that next notebook; first, meet $\rho$ for a state you know well, $|+\rangle$.

```python
rho_plus = density.density_matrix(KET_PLUS)
print("rho for |+> =\n", np.round(rho_plus, 3))
print("valid density matrix:", density.is_density_matrix(rho_plus))

qubit_basis = ["|0>", "|1>"]
viz.plot_density_matrix(rho_plus, title="rho for |+>", basis_labels=qubit_basis)
plt.show()
```

**Read the figure.** $\rho$ is drawn as two heatmaps — its real and imaginary parts (the imaginary part is all zero here). Every entry is $\tfrac12$. `is_density_matrix` confirms it passes the three tests every $\rho$ must satisfy: **Hermitian** (equal to its own conjugate transpose), **unit trace** (the diagonal sums to 1), and **positive** (no negative probabilities hiding inside).

## 2. Where the information lives

The two parts of $\rho$ mean different things, and telling them apart is the key skill of this whole module:

- the **diagonal** entries are the $Z$-measurement probabilities from notebook 03 — here $(0.5, 0.5)$, exactly $|+\rangle$'s coin flip;
- the **off-diagonal** entries are the **coherence** — the fingerprint of a genuine superposition, the part a $Z$ measurement was blind to.

```python
diag = np.real(np.diag(rho_plus))
coherence = abs(rho_plus[0, 1])
print("diagonal (Z-measurement probabilities):", np.round(diag, 3))
print("off-diagonal magnitude (coherence)    :", round(float(coherence), 3))
```

**Read the output.** The diagonal reproduces $|+\rangle$'s $50/50$ split — nothing a classical probability vector could not say. The off-diagonal $0.5$ is the **coherence**, and it is maximal for a pure equal superposition. In the very next notebook we meet a state with the *same diagonal* but a *zero* off-diagonal — and that one contrast is the seed of the entire course.
