# 06 — Pure vs Mixed

Meet the maximally mixed state, set it beside $|+\rangle$, and find the single contrast the whole course is built to explain.

**Prerequisites:** `01/05_from_states_to_density_matrix`.

**What you'll be able to do**
- Distinguish a superposition from a statistical mixture — two very different kinds of uncertainty.
- Build the maximally mixed state $I/2$ and compare it to $|+\rangle$.
- Measure "how mixed" a state is with purity and von Neumann entropy.

This is, quietly, the most important notebook in the foundations. You will build two states that look identical to a $Z$ measurement, yet are about as different as two states can be — and that difference is the exact reason the rest of the course exists. Take your time here; it rewards it.

```python
import numpy as np
import matplotlib.pyplot as plt

from qot_course import viz
from qot_course.quantum import density
from qot_course.quantum.states import KET_PLUS

np.random.seed(42)
viz.use_course_style()
```

## 1. Two kinds of uncertainty

$|+\rangle$ is uncertain in a *quantum* way: it is genuinely both $0$ and $1$ at once, in superposition. There is a second, completely different kind of uncertainty — the plain *classical* kind.

Imagine a friend flips a fair coin and hands you $|0\rangle$ on heads, $|1\rangle$ on tails, without telling you which. You are *classically uncertain*: the qubit is in one definite state, you do not know which one. That is a **mixed state**:

$$ \rho = \tfrac12|0\rangle\langle0| + \tfrac12|1\rangle\langle1| = \tfrac{I}{2}, $$

the **maximally mixed** state.

```python
rho_plus = density.density_matrix(KET_PLUS)
rho_mixed = density.maximally_mixed(2)
print("rho for I/2 =\n", np.round(rho_mixed, 3))

qubit_basis = ["|0>", "|1>"]
viz.plot_density_matrix(rho_mixed, title="rho for I/2 (maximally mixed)", basis_labels=qubit_basis)
plt.show()
```

**Read the figure.** $I/2$ is diagonal: $\tfrac12$ on each of $|0\rangle$ and $|1\rangle$, and *nothing* off-diagonal. It encodes "it is one of these two, fifty-fifty" — a classical coin, written as an operator.

## 2. Same diagonal, different soul

Now set the two states side by side. Watch the diagonals — and then watch the corners.

```python
viz.plot_density_matrix(rho_plus, title="|+> : a pure superposition", basis_labels=qubit_basis)
viz.plot_density_matrix(rho_mixed, title="I/2 : a classical mixture", basis_labels=qubit_basis)
plt.show()
```

**Read both figures together.** $|+\rangle$ and $I/2$ have the **same diagonal** $(0.5, 0.5)$: a $Z$ measurement returns $50/50$ for *both* — it cannot tell them apart. But $|+\rangle$ carries **off-diagonal coherences** of $\tfrac12$, and $I/2$ has none. The difference between "genuinely both at once" and "one of the two, unknown" lives entirely in those corners.

## 3. Measuring mixedness

Two numbers make the contrast precise. **Purity** $\operatorname{tr}(\rho^2)$ is $1$ for a pure state and drops toward $1/d$ as a state mixes. **Von Neumann entropy** $S(\rho) = -\operatorname{tr}(\rho\log\rho)$ is $0$ for a pure state and rises to its maximum for the maximally mixed one.

```python
for name, rho in [("|+>", rho_plus), ("I/2", rho_mixed)]:
    print(f"{name:4s}  diagonal={np.round(np.real(np.diag(rho)), 3)}  "
          f"purity={density.purity(rho):.3f}  "
          f"entropy={density.von_neumann_entropy(rho):.3f} bits")
```

**Read the output.** Identical diagonals — but **purity** $1.0$ versus $0.5$, and **entropy** $0$ versus $1$ bit. $|+\rangle$ is pure and perfectly definite (we know the state exactly); $I/2$ is maximally mixed and carries a full bit of classical ignorance.

## The seed of the whole course

Sit with this for a moment. $|+\rangle$ and $I/2$ are indistinguishable to a $Z$ measurement — same diagonal — yet one holds maximal coherence and the other none; one is pure, the other maximally mixed.

A **classical** description sees only the diagonal. It literally cannot tell these two states apart. A description that sees the **whole operator** — the off-diagonal corners too — can.

That gap is what the entire course is built to explore:

> **Classical optimal transport sees only the diagonal; quantum optimal transport sees the whole $\rho$.**

Every quantum advantage we eventually find will live in those off-diagonal corners. You are holding the seed of it right now.
