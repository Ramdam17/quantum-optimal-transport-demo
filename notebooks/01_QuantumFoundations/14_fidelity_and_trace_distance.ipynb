# 14 — Fidelity & Trace Distance

Pick up two honest yardsticks for "how different are two states?", and watch mixedness become a position inside the Bloch ball.

**Prerequisites:** `01/13_pure_vs_mixed`.

**What you'll be able to do**
- Locate pure and mixed states in the Bloch ball — surface, interior, centre.
- Compare any two states with fidelity ($1$ = identical) and trace distance ($0$ = identical).
- Read mixedness directly as how far a state sits from the surface.

In notebook 02 every *pure* state sat on the *surface* of the Bloch sphere. Mixed states fill in the inside — and the deeper in, the more mixed. Today we make that picture quantitative, with two distances you will reach for again and again.

```python
import numpy as np
import matplotlib.pyplot as plt

from qot_course import viz
from qot_course.quantum import density
from qot_course.quantum.states import KET_0, KET_PLUS

np.random.seed(42)
viz.use_course_style()
```

## 1. The Bloch ball

Notebook 02 placed every *pure* qubit on the *surface* of the Bloch sphere. Mixed states live **inside**: the more mixed a state, the closer to the centre, where the maximally mixed $I/2$ sits. The length of the Bloch vector is a direct readout of purity — full length for a pure state, zero at the centre.

```python
rho_plus = density.density_matrix(KET_PLUS)
rho_mixed = density.maximally_mixed(2)
half_mixed = 0.5 * rho_plus + 0.5 * rho_mixed   # halfway to the centre

viz.plot_bloch(rho_plus, title="pure |+>  (on the surface)")
viz.plot_bloch(half_mixed, title="partly mixed  (inside)")
viz.plot_bloch(rho_mixed, title="I/2  (at the centre)")
plt.show()
```

**Read the figures.** As we mix in classical uncertainty, the Bloch vector **shrinks inward** — from the surface (pure $|+\rangle$), through the interior (half-mixed), to the centre ($I/2$, zero length). "How deep inside the ball" is exactly "how mixed". The geometry you learned for pure states now holds the mixed ones too.

## 2. Two yardsticks for telling states apart

How different are two states? Two standard measures answer it:

- **Fidelity** $F(\rho, \sigma)$: $1$ when the states are identical, falling toward $0$ when they are perfectly distinguishable.
- **Trace distance** $T(\rho, \sigma) = \tfrac12\|\rho - \sigma\|_1$: $0$ when identical, up to $1$ when perfectly distinguishable.

They run in opposite directions and answer slightly different questions, so having both is useful. Let's tabulate them for our three familiar states.

```python
states = {"|0>": density.density_matrix(KET_0), "|+>": rho_plus, "I/2": rho_mixed}
names = list(states)
print(f"{'pair':12s}{'fidelity':>10s}{'trace dist':>12s}")
for i in range(len(names)):
    for j in range(i + 1, len(names)):
        a, b = names[i], names[j]
        f = density.fidelity(states[a], states[b])
        t = density.trace_distance(states[a], states[b])
        print(f"{a + ' vs ' + b:12s}{f:10.3f}{t:12.3f}")
```

**Read the table.** $|0\rangle$ vs $|+\rangle$ are partly distinguishable; the pairs involving $I/2$ sit in between. Notice that fidelity and trace distance always move in opposite directions — high fidelity goes with low trace distance — two views of the same "closeness". You now have a quantitative answer to "are these the same state?", which the next notebook leans on directly.
