In the last notebook you wrote a qubit as two complex amplitudes and noticed it was quietly indexed by two angles. Those angles carry a beautiful meaning: they place the qubit on the surface of a sphere. From here on, one picture holds every pure state at once.

```python
import numpy as np
import matplotlib.pyplot as plt

from qot_course import viz
from qot_course.quantum.states import KET_0, KET_1, KET_PLUS, qubit_state, bloch_vector

np.random.seed(42)
viz.use_course_style()
```

## 1. A sphere of states

Here is the central picture. Every **pure** qubit corresponds to exactly one point on the surface of a unit sphere — the **Bloch sphere**:

- the **north pole** is $|0\rangle$ and the **south pole** is $|1\rangle$ (the classical states),
- the **equator** holds the equal superpositions, like $|+\rangle$,
- every other point is some tilted, phase-shifted blend.

Let's drop our three known states onto it.

```python
viz.plot_bloch(KET_0, title="|0>  (north pole)")
viz.plot_bloch(KET_1, title="|1>  (south pole)")
viz.plot_bloch(KET_PLUS, title="|+>  (equator)")
plt.show()
```

**Read the figures.** $|0\rangle$ points straight up and $|1\rangle$ straight down — the two classical states sit at opposite poles, as far apart as the sphere allows. $|+\rangle$ lies flat on the equator, exactly between them: an equal superposition is, geometrically, the halfway point. Distance on this sphere is about to mean something physical, and you are already reading it.

## 2. Two angles, one point

The two angles you passed to `qubit_state` last time are exactly this sphere's coordinates:

$$ |\psi\rangle = \cos(\theta/2)\,|0\rangle + e^{i\phi}\sin(\theta/2)\,|1\rangle. $$

- $\theta$ (the polar angle) sets the **latitude** — how far down from the north pole, i.e. how much $|1\rangle$ is mixed in.
- $\phi$ (the azimuthal angle) sets the **longitude** — the phase, i.e. where around the equator the state points.

Let's place a tilted state and read its coordinates back out with `bloch_vector`.

```python
psi = qubit_state(theta=2 * np.pi / 3, phi=np.pi / 4)   # 120 deg down, 45 deg around
r = bloch_vector(psi)
print("Bloch vector (x, y, z) =", np.round(r, 3))
print("length =", round(float(np.linalg.norm(r)), 6), " (pure states sit on the surface)")

viz.plot_bloch(psi, title="|psi>  (a tilted state)")
plt.show()
```

**Read the figure and the output.** The tilted state points up-and-around: its latitude comes from $\theta = 120^\circ$ (well below the north pole, so a good deal of $|1\rangle$ is mixed in) and its longitude from $\phi = 45^\circ$ (a phase that swings it off the $x$-axis). Its Bloch vector has length exactly $1$ — every pure state lives on the **surface**. Hold that thought: in the density-matrix notebooks, the states *inside* the ball will turn out to be the genuinely mixed ones.
