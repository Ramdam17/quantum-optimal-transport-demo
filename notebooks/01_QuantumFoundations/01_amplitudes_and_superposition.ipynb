# 01 — Amplitudes & Superposition

Write a qubit as a pair of complex amplitudes and meet superposition — the first genuinely quantum idea.

**Prerequisites:** `00/01_environment_setup`.

**What you'll be able to do**
- Write a qubit as a two-component complex vector and read its amplitudes.
- Recognise the basis states $|0\rangle$, $|1\rangle$ and the equal superposition $|+\rangle$.
- Check the one rule every qubit must obey: its amplitudes are normalised.

A classical bit is the plainest thing in computing: it is 0, or it is 1. A qubit keeps both possibilities in play at once, each weighted by a number called an **amplitude**. That single move — from "one of two" to "a blend of two" — is where everything quantum begins, and you are about to write it down with your own hands.

```python
import numpy as np
import matplotlib.pyplot as plt

from qot_course import viz
from qot_course.colors import COLORS
from qot_course.quantum.states import KET_0, KET_1, KET_PLUS, qubit_state

np.random.seed(42)
viz.use_course_style()
```

## 1. From a bit to a qubit

A **qubit** is a unit vector in a two-dimensional complex space:

$$ |\psi\rangle = a_0\,|0\rangle + a_1\,|1\rangle, \qquad a_0, a_1 \in \mathbb{C}. $$

The numbers $a_0, a_1$ are the **amplitudes**. Two of them deserve names right away: $|0\rangle$ and $|1\rangle$ are the "classical" basis states (all the weight on one outcome), and

$$ |+\rangle = \tfrac{1}{\sqrt 2}\big(|0\rangle + |1\rangle\big) $$

puts **equal** amplitude on both — our first superposition. The course ships these as ready-made vectors.

```python
print("|0> =", np.round(KET_0, 3))
print("|1> =", np.round(KET_1, 3))
print("|+> =", np.round(KET_PLUS, 3))
```

**Read the output.** Each qubit is stored as two numbers — the amplitudes on $|0\rangle$ and on $|1\rangle$. $|0\rangle$ is `[1, 0]` (all the weight on the first slot), $|1\rangle$ is `[0, 1]`, and $|+\rangle$ is `[0.707, 0.707]` — the same weight, $1/\sqrt2$, on each. That even split *is* superposition.

## 2. Seeing the split

Let's picture how each state shares its amplitude between the two slots. The basis states put everything in one bar; $|+\rangle$ shares equally.

```python
states = {"|0>": KET_0, "|1>": KET_1, "|+>": KET_PLUS}
palette = [COLORS["source"], COLORS["target"], COLORS["quantum"]]

fig, ax = plt.subplots(figsize=(7.5, 4.2))
x = np.arange(2)
width = 0.26
for k, (name, s) in enumerate(states.items()):
    ax.bar(x + (k - 1) * width, np.real(s), width=width, color=palette[k], label=name)
ax.set_xticks(x)
ax.set_xticklabels([r"amplitude on $|0\rangle$", r"amplitude on $|1\rangle$"])
ax.set_ylabel("amplitude")
ax.set_title("Superposition: how each state splits its amplitude", pad=12)
ax.legend()
plt.show()
```

**Read the figure.** $|0\rangle$ (periwinkle) puts all its amplitude on the left bar, $|1\rangle$ (amber) all on the right. $|+\rangle$ (sky blue) splits its amplitude evenly across both — it is genuinely *both at once*, not secretly one or the other. Holding two possibilities together like this is the resource the whole course will build on.

## 3. The one rule: normalisation

Amplitudes are not free to be anything. A qubit must be a **unit** vector: the squared sizes of its amplitudes add up to one,

$$ |a_0|^2 + |a_1|^2 = 1. $$

To see this on a less tidy state, build a *tilted* qubit. The course's `qubit_state` takes two angles $(\theta, \phi)$ and returns $\cos(\theta/2)\,|0\rangle + e^{i\phi}\sin(\theta/2)\,|1\rangle$. Do not worry about what the angles *mean* geometrically yet — that is the next notebook's gift. For now, confirm the rule holds.

```python
psi = qubit_state(theta=2 * np.pi / 3, phi=np.pi / 4)
print("a tilted state |psi> =", np.round(psi, 3))
print("amplitudes:  a0 =", np.round(psi[0], 3), " a1 =", np.round(psi[1], 3))
print("|a0|^2 + |a1|^2 =", round(float(np.abs(psi[0]) ** 2 + np.abs(psi[1]) ** 2), 6))
```

**Read the output.** The amplitude on $|1\rangle$ is now a *complex* number — it carries a **phase** ($e^{i\phi}$). Yet the squared sizes still sum to exactly $1$. Every valid qubit obeys this: the total "amount of state" is conserved, no matter how you tilt it. The two angles $(\theta, \phi)$ quietly index a whole sphere of states — which is exactly the picture we draw next.
