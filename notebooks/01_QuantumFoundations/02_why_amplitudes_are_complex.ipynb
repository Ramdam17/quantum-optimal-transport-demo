```python
import numpy as np
import matplotlib.pyplot as plt

from qot_course import viz
from qot_course.colors import COLORS
from qot_course.quantum.states import KET_PLUS, KET_MINUS

np.random.seed(42)
viz.use_course_style()
```

## 1. Two states classical weights cannot tell apart

Here are two superpositions. Each puts the *same* amount of weight on $|0\rangle$ and on $|1\rangle$:

$$ |+\rangle = \tfrac{1}{\sqrt2}\big(|0\rangle + |1\rangle\big), \qquad |-\rangle = \tfrac{1}{\sqrt2}\big(|0\rangle - |1\rangle\big). $$

The only difference is a minus sign on the second amplitude. Let's look at the amplitudes and at their squared sizes (the *weights*).

```python
for name, state in [("|+>", KET_PLUS), ("|->", KET_MINUS)]:
    print(f"{name}: amplitudes = {np.round(state, 3)}   weights |a|^2 = {np.round(np.abs(state)**2, 3)}")
```

**Read the output.** Both states carry weight $0.5$ on each outcome — identical, as far as a list of probabilities is concerned. A classical description ("50% this, 50% that") *cannot* tell $|+\rangle$ from $|-\rangle$. Yet they are genuinely different physical states. The difference lives entirely in the **sign** of the second amplitude — and a sign is the simplest possible phase. To hold that information, amplitudes need more than a non-negative size.

## 2. Interference: the reason a sign matters

A sign matters because amplitudes **add**, and when they add they can *cancel*. Probabilities (non-negative reals) can only pile up; amplitudes can subtract. Watch what happens when we combine $|0\rangle$ and $|1\rangle$ with a relative phase $\varphi$ and ask for the weight of the combination:

$$ s(\varphi) = \tfrac{1}{2}\big(1 + e^{i\varphi}\big), \qquad |s(\varphi)|^2 = \cos^2(\varphi/2). $$

At $\varphi=0$ the two contributions reinforce (weight $1$); at $\varphi=\pi$ they perfectly cancel (weight $0$). That cancellation is **destructive interference** — and it is impossible with non-negative weights.

```python
phi = np.linspace(0, 2 * np.pi, 400)
amplitude_sum = 0.5 * (1 + np.exp(1j * phi))
weight = np.abs(amplitude_sum) ** 2

fig, ax = plt.subplots(figsize=(7.5, 4.2))
ax.plot(phi, weight, color=COLORS["quantum"], lw=2.5)
ax.axhline(0, color=COLORS["grid"], lw=1)
ax.scatter([0, np.pi], [1, 0], color=COLORS["highlight"], zorder=5, s=60)
ax.annotate("constructive", (0, 1), textcoords="offset points", xytext=(10, -14), color=COLORS["text"])
ax.annotate("destructive", (np.pi, 0), textcoords="offset points", xytext=(10, 12), color=COLORS["text"])
ax.set_xlabel(r"relative phase $\varphi$")
ax.set_ylabel(r"weight $|s(\varphi)|^2$")
ax.set_title("Interference: a relative phase steers the outcome", pad=12)
plt.show()
```

**Read the figure.** Turning the relative phase $\varphi$ smoothly moves the combined weight from $1$ (full reinforcement) down to $0$ (full cancellation) and back. Nothing about the *sizes* of the two amplitudes changed — only their phase relationship. This dial is real, and it is the engine behind every quantum algorithm. Real, non-negative numbers have no such dial; complex numbers do.

## 3. Polar form: magnitude and phase

Every complex amplitude splits cleanly into a size and an angle (Euler's formula):

$$ a = r\,e^{i\varphi}, \qquad r = |a| \ge 0, \quad \varphi = \arg(a) \in (-\pi, \pi]. $$

The magnitude $r$ sets the *weight* $r^2$ (how likely that outcome is — the Born rule, made precise in `01/04`). The angle $\varphi$ is the new degree of freedom that carries the interference we saw above.

```python
a = 0.612 + 0.612j          # the |1> amplitude of the tilted state from 01/01
r, phi = np.abs(a), np.angle(a)
print(f"a = {a}")
print(f"magnitude r = {r:.3f}   ->   weight r^2 = {r**2:.3f}")
print(f"phase phi   = {phi:.3f} rad = {np.degrees(phi):.1f} deg")
print(f"reconstructed r*e^(i*phi) = {np.round(r * np.exp(1j * phi), 3)}")
```

**Read the output.** The single complex number $a$ decomposes into a magnitude ($\approx 0.866$, so weight $\approx 0.75$) and a phase ($45^\circ$). Rebuilding $r\,e^{i\varphi}$ returns the original amplitude exactly. Let's picture the phase as an angle in the complex plane.

```python
fig, ax = plt.subplots(figsize=(5.2, 5.2))
circle = plt.Circle((0, 0), 1.0, fill=False, color=COLORS["grid"], lw=1.5)
ax.add_artist(circle)
for ang, col, lab in [(np.pi / 4, COLORS["quantum"], r"$\varphi=45°$"),
                      (2 * np.pi / 3, COLORS["highlight"], r"$\varphi=120°$")]:
    ax.annotate("", xy=(np.cos(ang), np.sin(ang)), xytext=(0, 0),
                arrowprops=dict(arrowstyle="->", color=col, lw=2.5))
    ax.text(1.1 * np.cos(ang), 1.1 * np.sin(ang), lab, color=col, ha="center")
ax.axhline(0, color=COLORS["grid"], lw=1)
ax.axvline(0, color=COLORS["grid"], lw=1)
ax.set_xlim(-1.3, 1.3); ax.set_ylim(-1.3, 1.3)
ax.set_aspect("equal")
ax.set_xlabel("real part"); ax.set_ylabel("imaginary part")
ax.set_title("An amplitude of unit size is a point on a circle", pad=12)
plt.show()
```

**Read the figure.** A unit-magnitude amplitude is a point on the unit circle, and its phase is the angle it makes. Changing the phase slides the arrow around the circle without changing its length — exactly the "dial" of section 2. This circle is a preview: in `01/03` the relative phase between $|0\rangle$ and $|1\rangle$ becomes the *longitude* on the Bloch sphere.

## 4. Global phase is free; relative phase is physical

One subtlety keeps everyone honest. Multiply an *entire* state by a phase $e^{i\alpha}$ — a **global phase** — and every measurement weight $|\langle x|\psi\rangle|^2$ is unchanged, because $|e^{i\alpha}|=1$. A global phase is physically invisible; you may add or drop one at will.

What you may *not* drop is a **relative** phase — a phase on one amplitude but not the other. That is the difference between $|+\rangle$ and $|-\rangle$, and it changes interference.

```python
alpha = np.pi / 3
plus_global = np.exp(1j * alpha) * KET_PLUS          # |+> with a global phase
print("weights of |+>            :", np.round(np.abs(KET_PLUS)**2, 3))
print("weights of e^{ia}|+>      :", np.round(np.abs(plus_global)**2, 3), " (global phase: identical)")
print("weights of |->            :", np.round(np.abs(KET_MINUS)**2, 3), " (relative phase: also 0.5/0.5...)")
print()
print("overlap <+|->             :", np.round(np.vdot(KET_PLUS, KET_MINUS), 3), " (= 0: orthogonal, genuinely different)")
print("overlap <+|e^{ia}+>       :", np.round(np.abs(np.vdot(KET_PLUS, plus_global)), 3), " (= 1: same physical state)")
```

**Read the output.** A global phase leaves every weight untouched *and* keeps the state physically identical (overlap magnitude $1$ with the original). The relative phase that turns $|+\rangle$ into $|-\rangle$ also leaves the weights at $0.5/0.5$ — but those two states are *orthogonal* (overlap $0$): as different as $|0\rangle$ and $|1\rangle$. Weights alone miss this entirely; the phase is where the difference lives. (We make "overlap" precise in `01/05`.)
