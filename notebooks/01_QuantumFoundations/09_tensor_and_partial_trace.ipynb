# 09 — Tensor Product & Partial Trace

Combine two qubits into one system, then recover a part — meeting the quantum version of a marginal.

**Prerequisites:** `01/05_from_states_to_density_matrix`.

**What you'll be able to do**
- Combine two systems with the tensor (Kronecker) product.
- Build the joint density matrix of a two-qubit state.
- Take a partial trace and recognise it as the quantum marginal.

So far, one qubit. The genuinely interesting quantum phenomena — correlation, entanglement, the inter-system structure that quantum optimal transport ultimately compares — need at least two. This notebook sets up the two-system stage and hands us the tool for asking what system A is doing on its own.

```python
import numpy as np
import matplotlib.pyplot as plt

from qot_course import viz
from qot_course.quantum.states import KET_0, KET_PLUS
from qot_course.quantum.density import density_matrix
from qot_course.quantum.composite import tensor, partial_trace

np.random.seed(42)
viz.use_course_style()
```

## 1. Two systems: the tensor product

To describe two qubits together, combine their spaces with the **tensor product** $\mathcal{H}_A \otimes \mathcal{H}_B$, of dimension $2 \times 2 = 4$. States and operators combine through the **Kronecker product**. Let's build the simple product state $|0\rangle \otimes |+\rangle$ — qubit A in $|0\rangle$, qubit B in $|+\rangle$.

```python
state_AB = tensor(KET_0, KET_PLUS)
rho_AB = density_matrix(state_AB)
print("composite state dimension:", state_AB.shape[0])

two_qubit_basis = ["|00>", "|01>", "|10>", "|11>"]
viz.plot_density_matrix(
    rho_AB,
    title="rho for |0> tensor |+>  (a product state)",
    basis_labels=two_qubit_basis,
)
plt.show()
```

**Read the figure.** A $4\times4$ density matrix, indexed by the basis $|00\rangle, |01\rangle, |10\rangle, |11\rangle$. All the mass sits in the top-left block because qubit A is $|0\rangle$; inside that block, the $|+\rangle$ coherences of qubit B appear. The joint state is the two single-qubit states placed side by side.

## 2. The partial trace = the quantum marginal

Recall the classical **marginal**: sum a joint distribution over one variable to recover the other's distribution on its own. The quantum version is the **partial trace**: $\rho_A = \operatorname{tr}_B(\rho_{AB})$ gives the state of qubit A alone. Let's recover both factors from our product state.

```python
reduced_A = partial_trace(rho_AB, keep=[0], dims=[2, 2])
reduced_B = partial_trace(rho_AB, keep=[1], dims=[2, 2])
print("reduced A == |0>? ", np.allclose(reduced_A, density_matrix(KET_0)))
print("reduced B == |+>? ", np.allclose(reduced_B, density_matrix(KET_PLUS)))
```

**Read the output.** Tracing out B recovers A *exactly* (and tracing out A recovers B) — for a product state, each part is perfectly well-defined, exactly as marginalising independent variables would give. Hold onto the word *independent*: in the next notebook we meet a two-qubit state whose parts are not well-defined at all, and that is where things turn genuinely quantum.
