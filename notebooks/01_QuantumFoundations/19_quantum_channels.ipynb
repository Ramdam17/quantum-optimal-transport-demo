## 1. Quantum channels: noisy evolution

The quantum version of a **Markov kernel** is a **quantum channel** — a completely-positive, trace-preserving (CPTP) map

$$ \rho \mapsto \sum_k K_k\,\rho\,K_k^\dagger, \qquad \sum_k K_k^\dagger K_k = I. $$

The matrices $K_k$ are its **Kraus operators**; the completeness relation on the right is what guarantees probabilities still sum to one. The **depolarizing channel** mixes a state toward $I/2$. Let's push $|+\rangle$ through increasing noise and track its purity.

```python
print("depolarizing channel is CPTP:", is_cptp(depolarizing_channel(0.3)))

ps = np.linspace(0.0, 1.0, 11)
rho_plus = density_matrix(KET_PLUS)
purities = [purity(apply_channel(rho_plus, depolarizing_channel(p))) for p in ps]

fig, ax = plt.subplots(figsize=(8, 4.5))
ax.plot(ps, purities, "-o", color=viz.FLOW_COLOR, markersize=8)
ax.axhline(0.5, color=viz.TARGET_COLOR, linestyle="--", label="I/2 (fully mixed)")
ax.set_xlabel("depolarizing parameter p")
ax.set_ylabel("purity  tr(rho$^2$)")
ax.set_title("A quantum channel mixes a state toward I/2", pad=12)
ax.legend()
plt.show()
```

**Read the figure.** As the noise $p$ grows, purity falls from $1$ (the pure $|+\rangle$) toward $0.5$ (the maximally mixed $I/2$). This is exactly the noise that turned notebook 08's tomographed $|+\rangle$ into a *mixed* state — and now we can name it: a quantum channel acting on the qubit. Decoherence is not mysterious; it is a CPTP map, and you can watch it work.
