## 2. The greatest hits — every promise the course made, delivered

A course can claim a great deal. This one can show its work. The `course_greatest_hits()` function in `qot_course.quantum_ot.synthesis` re-runs the canonical numerical demonstrations of the course — twenty results spanning information theory, information geometry, classical transport, and the quantum lift — in a single call. Every entry is a promise some notebook made, recomputed from the validated source so you can watch it land.

We present them **grouped by the module that earned each one**, so the table reads as the course's spine rather than a flat dump of numbers. The values come straight from `course_greatest_hits()`; the grouping is ours, mapping each result back to where you met it.


```python
# Re-run every canonical result of the course (a few seconds: this calls the real
# pipeline -- the SDP / Sinkhorn solvers print accuracy notes, which we leave visible).
hits = course_greatest_hits()

# The function's dict keys carry the course's old per-session shorthand as a prefix
# (a leading token of the form "<letter><number>:"). We never display that shorthand:
# we read the leading NUMBER off each key and map it to the MODULE that earned the
# result, then present a clean, human-readable label. (We read the function's keys and
# values; we relabel only what is shown -- the src function is untouched.)
import re as _re

MODULE_INFO_GEOMETRY = "02 - Information and Geometry"
MODULE_CLASSICAL_OT = "03 - Classical Optimal Transport"
MODULE_QUANTUM_OT = "04 - Quantum Optimal Transport"
MODULE_ORDER = [MODULE_INFO_GEOMETRY, MODULE_CLASSICAL_OT, MODULE_QUANTUM_OT]


def _module_for_index(n: int) -> str:
    """Map the leading index of a greatest-hits key to the module that earned it.

    The course's information / geometry results carry the lowest indices, the classical
    optimal-transport results the middle band, and the quantum lift the highest -- the
    same five-module ordering the course is taught in.
    """
    if n <= 7:
        return MODULE_INFO_GEOMETRY
    if n <= 11:
        return MODULE_CLASSICAL_OT
    return MODULE_QUANTUM_OT


def _clean_label(label: str) -> str:
    """Rewrite an in-text cross-reference in the old shorthand to its module name.

    One result's label cross-references an earlier one in the old per-session shorthand
    ("... from <old-tag>"); we rewrite that single embedded reference to the module it
    lived in, so NOTHING displayed uses the old labels. We target only the cross-reference
    pattern ("from <letter><number>") so that genuine math symbols such as W2 (the
    Wasserstein-2 distance) are left untouched.
    """
    return _re.sub(
        r"from\s+[A-Za-z]([0-9]{1,2})\b",
        lambda m: "from Module " + _module_for_index(int(m.group(1))).split(" - ")[0],
        label,
    )


def _module_and_label(raw_key: str) -> tuple[str, str]:
    """Map a raw greatest-hits key to (module, clean label) without the old shorthand."""
    prefix, _, rest = raw_key.partition(":")
    # Read the leading index off the prefix (a few keys are compound, e.g. "<tag>/<tag>").
    lead_index = int(_re.search(r"[0-9]+", prefix).group(0))
    return _module_for_index(lead_index), _clean_label(rest.strip())


# Group the results by module, preserving the function's within-module order.
grouped: dict[str, list[tuple[str, float]]] = {m: [] for m in MODULE_ORDER}
for raw_key, value in hits.items():
    module, label = _module_and_label(raw_key)
    grouped.setdefault(module, []).append((label, value))

# Print as a module-grouped table -- the course's promises, delivered, by movement.
width = max(len(lbl) for entries in grouped.values() for lbl, _ in entries)
for module in MODULE_ORDER:
    print(f"\nModule {module}")
    print("-" * (width + 18))
    for label, value in grouped[module]:
        print(f"  {label:<{width}s}  {value:>12.6f}")

print(f"\n{len(hits)} canonical results, every one recomputed live from the validated source.")
```

**Read the table.** Every line is a promise some notebook made, recomputed here from scratch — and grouped by the module that earned it.

The two-geometries-on-one-simplex punchline shows in two numbers: the mixture midpoint puts almost no mass in the middle ($\approx 0$), while the Wasserstein midpoint moves the whole bump there (mass $\approx 1$). Then the quantum surprises that classical information theory forbids: the Bell pair's quantum mutual information reaches **2 bits** — twice the classical ceiling — and its conditional entropy is **$-1$ bit**, genuinely negative. And the seed of the whole course: $|+\rangle\langle+|$ and $I/2$ share a $Z$-diagonal yet differ, with relative entropy $1$ bit and a strictly positive Bures distance.

**Module 03 — classical optimal transport.** Where Monge's deterministic map fails, the Kantorovich linear program still returns a finite cost. The one-dimensional Wasserstein closed form matches the linear-program value to machine precision — the quantile gift, "sort and match". And the bridge to the quantum world is already visible: the square root of the Bures–Wasserstein matrix term on density matrices equals the Bures distance from Module 02, to the digit. ## 3. The headline result — quantum optimal transport beats the phase-locking value

Of everything the course delivered, one result is the reason the capstone was written, and it deserves the spotlight here at the close.

The question was deliberately hard, and the course refused to answer it cheaply: **can a quantum-optimal-transport coupling measure do something the phase-locking value (PLV) cannot?** The phase-locking value is the workhorse of phase-synchrony analysis — the magnitude of the first circular moment of the phase difference, $\mathrm{PLV} = |\langle e^{i\Delta\theta}\rangle|$. To "beat" it means to read a coupling it is blind to, on a system where it is genuinely blind. The capstone earned the answer in three decisive moves (`13`, `14`, `15`) and reckoned with it honestly (`16`).

**The redundancy reveal (`13`).** First, the uncomfortable truth: on the naive phase-qubit embedding, the quantum measures were PLV *in disguise*. The rank correlation $\mathrm{Spearman}(\mathrm{PLV}, \mathrm{QMI}) \approx 1$, and the operator showed exactly why — the only coupling-bearing element of $\rho_{AB}$ was an off-block coherence with $|\rho_{AB}[1,2]| = \mathrm{PLV}/4$. A state whose only moving part *is* PLV can feed a measure nothing else. This was a property of the *embedding*, not of quantum optimal transport — which told us precisely what to change.

**The richer embedding (`14`).** Swapping the qubit for the multi-frequency qutrit $\tfrac{1}{\sqrt 3}\bigl(|0\rangle + e^{i\theta}|1\rangle + e^{2i\theta}|2\rangle\bigr)$ gave $\rho_{AB}$ a *second* coupling channel — the element $\rho_{AB}[6,2]$, carrying the **second** circular moment, a degree of freedom independent of PLV. The redundancy was broken in principle: two couplings can share a PLV yet differ in that second moment.

**The discriminating experiment (`15`).** Then the structural opening became a *measured* advantage. Two ensembles were built with identical PLV by construction (matched first moment $a_1 = 0.4$) but a different second moment ($a_2 = 0.0$ versus $0.3$). PLV could not tell them apart; the naive embedding's quantum mutual information could not either (exactly as `13` predicted); and the **richer** embedding's quantum mutual information *separated* them — by a gap of about $0.004$ nats, standing several times above an $a_2$-matched null across many random seeds. **That is the headline the restructure exists to deliver: a controlled, significant case where quantum optimal transport beats the phase-locking value.**

And then — because a result you can trust is one whose limits you can also state — the honest accounting (`16`). The advantage is real, and it has a bill. It appears *only* when the coupling carries higher-order structure the embedding captures (never on the naive embedding, never for purely first-moment coupling); the effect is small and was clean only because $N = 150{,}000$ samples per channel pushed the Monte-Carlo error below the gap; and — stated plainly so it cannot be smuggled back — **a purely classical higher-order statistic** (a second-harmonic PLV $|\langle e^{2i\Delta\theta}\rangle|$, or the bispectrum) **would also separate the two ensembles.** There is no quantum magic. The genuine value of quantum optimal transport here is the **unified framework**: one transport / information-geometric language placing classical and quantum coupling side by side, reaching higher-order structure *through the embedding choice* rather than through a statistic hand-picked in advance. Choose a richer embedding, and the same two measures — quantum mutual information and the Bures distance — automatically read whatever structure that embedding now carries.

Hold both halves at once. The course set out to show, not assert, when this machinery earns its keep — and it did: a real separation where PLV is blind, priced honestly and located precisely. That balance, a celebrated result and its candid bound in the same hand, is the course's culminating achievement.

| Classical | Quantum | Earned in |
|-----------|---------|-----------|
| probability vector $a \in \Delta^n$ | density matrix $\rho \in \mathcal{S}(\mathcal{H})$ | Module 01 |

| marginal | partial trace | Module 01 |
| product distribution $p_X p_Y$ | product state $\rho_A \otimes \rho_B$ | Module 01 |
| Markov kernel | quantum channel (CPTP map) | Module 01 |
| Shannon entropy $H(p)$ | von Neumann entropy $S(\rho)$ | Modules 01–02 |
| KL divergence $D(p\,\|\,q)$ | Umegaki relative entropy $S(\rho\,\|\,\sigma)$ | Module 02 |

| conditional entropy $H(X\lvert Y) \ge 0$ | quantum conditional entropy $S(A\lvert B)$ — **can be negative** | Module 02 |
| Fisher–Rao metric | Bures / quantum Fisher information | Module 02 |
| transport plan / coupling $P \in T(a, b)$ | **quantum coupling $\rho_{AB} \succeq 0$ with marginals** | Module 04 |
| Kantorovich LP $\min \langle C, P\rangle$ | **quantum OT SDP $\min \mathrm{tr}(C\,\rho_{AB})$** | Module 04 |
| transportation polytope $T(a, b)$ | **spectrahedron of quantum couplings** | Module 04 |
| Wasserstein-$p$ distance $W_p$ | **quantum Wasserstein** (several non-equivalent definitions) | Module 04 |

| $W_1 = \int \lvert F_\mu - F_\nu\rvert\,\mathrm{d}x$ | qubit-$W_1$: Lipschitz dual, no integral form | Modules 03–04 |
| Gaussian Bures–Wasserstein $W_2$ | Bures distance on density matrices (the bridge) | **Module 03** |
| McCann displacement geodesic | quantum displacement / Carlen–Maas dynamic | Module 04 |
| Sinkhorn = KL projection of $K = e^{-C/\varepsilon}$ | **quantum Sinkhorn = Umegaki projection of $K = \exp(-C/\varepsilon)$** | Module 04 |
| Benamou–Brenier / Otto Riemannian metric | Carlen–Maas quantum Riemannian metric | Modules 03–04 |
| phase-locking value $\mathrm{PLV} = \lvert\langle e^{i\Delta\theta}\rangle\rvert$ | first-moment coherence of $\rho_{AB}$ ($\lvert\rho_{AB}[1,2]\rvert = \mathrm{PLV}/4$, naive embedding) | **Module 04 capstone** |

| Birkhoff polytope / Birkhoff–von Neumann | (analogous structural result open) | Modules 03–04 |

This dictionary is the *spine of the course*. Every row was earned with code, a derivation, and a "Read the figure" explanation — and read together, the bottom three rows are the capstone's punchline in compact form: PLV is the first-moment special case of a single geometric coupling measure whose reach is set, in one place, by the embedding. The honest blanks — transfer entropy, a Birkhoff–von Neumann analogue, a one-dimensional operator quantile — are not gaps in the course but acknowledgements of where the analogy genuinely breaks down, and where research opens up.
