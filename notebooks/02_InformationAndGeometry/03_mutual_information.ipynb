# 03 — Mutual Information

Measure how much two variables share — the information one carries about the other.

**Prerequisites:** `02/02_kl_divergence`.

**What you'll be able to do**
- Define mutual information $I(X;Y)$ as the KL divergence between the joint and the product of marginals.
- Compute $I(X;Y)$ from a 2-D joint distribution.
- See it rise from $0$ (independent) to its maximum (perfectly correlated).

KL compared a model to the truth. Turn it inward and it answers a new question: how far is the *joint* distribution of two variables from the case where they are independent? That gap is the information they share — mutual information — the workhorse of coupling analysis and the classical cousin of the quantum mutual information you will meet in `02/10`.

```python
import numpy as np
import matplotlib.pyplot as plt

from qot_course import viz
from qot_course.infotheory.classical import mutual_information

np.random.seed(42)
viz.use_course_style()
```

## 1. Shared information

The **mutual information** of two variables is the KL divergence between their joint distribution $p_{XY}$ and the product of their marginals $p_X p_Y$:

$$ I(X;Y) = D\!\left(p_{XY}\,\middle\|\,p_X p_Y\right) = \sum_{x,y} p(x,y)\log_2\frac{p(x,y)}{p(x)\,p(y)}. $$

It is **zero iff $X$ and $Y$ are independent** (then the joint *is* the product), and it grows as they become more dependent. Equivalently $I(X;Y) = H(X) + H(Y) - H(X,Y)$: the bits you save by describing the two together rather than apart.

```python
independent = np.outer([0.5, 0.5], [0.5, 0.5])
correlated = np.array([[0.5, 0.0], [0.0, 0.5]])
print(f"I(X;Y)  independent          = {mutual_information(independent):.3f} bits")
print(f"I(X;Y)  perfectly correlated = {mutual_information(correlated):.3f} bits")
```

**Read the output.** Independent variables share $0$ bits — knowing $X$ tells you nothing about $Y$. Perfectly correlated variables share $1$ bit — knowing $X$ fixes $Y$ entirely. Everything interesting lives in between.

## 2. From independent to correlated

Blend the independent joint with the perfectly-correlated one and watch the mutual information climb.

```python
ts = np.linspace(0.0, 1.0, 50)
mi_curve = [mutual_information((1 - t) * independent + t * correlated) for t in ts]

fig, ax = plt.subplots(figsize=(8, 4.5))
ax.plot(ts, mi_curve, color=viz.SOURCE_COLOR, lw=2)
ax.set_xlabel("coupling  (0 = independent, 1 = perfectly correlated)")
ax.set_ylabel("mutual information  I(X;Y)  (bits)")
ax.set_title("Mutual information measures shared structure", pad=12)
plt.show()
```

**Read the figure.** $I(X;Y)$ rises smoothly from $0$ (independent) to $1$ bit (perfectly correlated). It is the amount of uncertainty about one variable that is removed by learning the other — a symmetric measure of how tightly the two are coupled.
