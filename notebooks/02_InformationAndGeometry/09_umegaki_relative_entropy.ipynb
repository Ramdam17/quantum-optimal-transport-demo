# 09 — Umegaki Relative Entropy

Lift the KL divergence to density matrices — and watch it see the coherence a classical KL is blind to.

**Prerequisites:** `01/13_pure_vs_mixed` (von Neumann entropy, the seed), `02/02_kl_divergence`.

**What you'll be able to do**
- Compute the Umegaki relative entropy $S(\rho\|\sigma) = \operatorname{tr}\rho(\log\rho - \log\sigma)$.
- See it distinguish $|+\rangle$ from $I/2$ — $1$ bit — though their $Z$-measurement statistics are identical.
- Recognise its asymmetry and its $+\infty$ on disjoint support, mirroring the classical KL.

Movement 2 began on the classical side: entropy, KL, mutual information. Now we lift them to density matrices, where two effects appear with no classical analogue. We start with the quantum KL — the Umegaki relative entropy — and it pays off immediately: it revives the $|+\rangle$ vs $I/2$ seed from `01/13` and sees exactly the coherence a classical KL throws away.

```python
import numpy as np
import matplotlib.pyplot as plt

from qot_course import viz
from qot_course.colors import COLORS
from qot_course.infotheory.quantum import quantum_relative_entropy
from qot_course.quantum.density import density_matrix, maximally_mixed, von_neumann_entropy
from qot_course.quantum.states import KET_0, KET_1, KET_PLUS

np.random.seed(42)
viz.use_course_style()
```

## 1. A one-line recap: von Neumann entropy

From `01/13`: the von Neumann entropy $S(\rho) = -\operatorname{tr}(\rho\log\rho)$ generalises Shannon entropy — $0$ for a pure state, $1$ bit for the maximally mixed qubit. We build the quantum KL on top of it.

```python
for name, rho in [("|0>", density_matrix(KET_0)),
                  ("|+>", density_matrix(KET_PLUS)),
                  ("I/2", maximally_mixed(2))]:
    print(f"S({name:4s}) = {von_neumann_entropy(rho):.3f} bit")
```

## 2. The quantum KL: Umegaki relative entropy

The quantum analogue of $D(p\|q)$ is the **Umegaki relative entropy**

$$ S(\rho\,\|\,\sigma) = \operatorname{tr}\!\left[\rho(\log\rho - \log\sigma)\right]. $$

It is non-negative, zero iff $\rho = \sigma$, asymmetric, and $+\infty$ whenever $\rho$ has support outside $\sigma$'s — every property of the classical KL, lifted.

**The punchline.** $|+\rangle\langle+|$ (pure) and $I/2$ (maximally mixed) have the *same* $Z$-diagonal $(1/2, 1/2)$. The classical KL between those diagonals is $0$. But Umegaki *sees the off-diagonal coherence* and returns $1$ bit — exactly the gap you first met in `01/13`.

```python
plus = density_matrix(KET_PLUS)
mixed = maximally_mixed(2)

s_quantum = quantum_relative_entropy(plus, mixed)
print(f"classical KL on Z-diagonals  D((1/2,1/2) || (1/2,1/2)) = 0.000 bit")
print(f"quantum     S(|+><+| || I/2)                           = {s_quantum:.3f} bit")
print()
print(f"asymmetry:  S(I/2 || |+><+|)    = {quantum_relative_entropy(mixed, plus)}   (support leaks -> inf)")
print(f"self:       S(|+><+| || |+><+|) = {quantum_relative_entropy(plus, plus):.3f}")
print(f"disjoint:   S(|0><0| || |1><1|) = {quantum_relative_entropy(density_matrix(KET_0), density_matrix(KET_1))}")

fig, ax = plt.subplots(figsize=(6.8, 4.5))
ax.bar(["classical KL\n(Z-diagonals)", "quantum Umegaki\nS(|+> || I/2)"],
       [0.0, s_quantum], color=[COLORS["source"], COLORS["highlight"]], alpha=0.9)
ax.set_ylabel("relative entropy (bits)")
ax.set_title("The quantum KL sees the coherence the classical KL misses", pad=12)
plt.show()
```

**Read the figure and output.** The classical KL on the diagonals is exactly $0$ — by that measure the two states are identical. The Umegaki relative entropy is $1$ bit — it sees the coherence. The asymmetry shows too: $S(I/2 \,\|\, |+\rangle\langle+|)$ is $+\infty$ because $I/2$ has support outside the rank-1 $|+\rangle\langle+|$, and disjoint supports give $+\infty$, exactly as the classical KL does. This relative entropy is the parent quantity of all of Movement 4 — entropic-regularised quantum optimal transport is a quantum KL projection.
