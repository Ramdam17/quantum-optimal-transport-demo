# 02 — KL Divergence

Measure the cost of a wrong model — the "extra bits" of using $q$ when the truth is $p$. This is the parent quantity of the entire course.

**Prerequisites:** `02/01_shannon_entropy`.

**What you'll be able to do**
- Compute the Kullback–Leibler divergence $D(p\|q)$ and read it as extra bits.
- Show KL is asymmetric ($D(p\|q) \neq D(q\|p)$) and zero only when $p = q$.
- See why KL — a divergence, not a symmetric distance — is the quantity Sinkhorn and quantum optimal transport are built on.

Entropy measured one distribution. Now we compare two. The KL divergence asks: if the data really come from $p$ but you model them as $q$, how many extra bits do you waste? That question — and its asymmetry — turns out to be the seed of entropic optimal transport, so it earns its own notebook.

```python
import numpy as np
import matplotlib.pyplot as plt

from qot_course import viz
from qot_course.infotheory.classical import kl_divergence

np.random.seed(42)
viz.use_course_style()
```

## 1. Extra bits: the KL divergence

How many *extra* bits do you pay by modelling data drawn from $p$ as if it came from $q$? That is the **Kullback–Leibler divergence**

$$ D(p\|q) = \sum_x p(x)\log_2\frac{p(x)}{q(x)} \ge 0. $$

It is zero exactly when $p = q$ and — crucially — it is **not symmetric**: $D(p\|q) \neq D(q\|p)$ in general. It is a *divergence*, not a distance.

```python
p = np.array([0.7, 0.2, 0.1])
q = np.array([0.4, 0.4, 0.2])
print(f"D(p || q) = {kl_divergence(p, q):.3f} bits")
print(f"D(q || p) = {kl_divergence(q, p):.3f} bits   (different — order matters)")
print(f"D(p || p) = {kl_divergence(p, p):.3f} bits   (zero iff identical)")
```

**Read the output.** $D(p\|q) \neq D(q\|p)$ — the order matters — and $D(p\|p) = 0$. KL is the **parent quantity** of this course: entropic-regularised optimal transport (Sinkhorn, in M3) is a KL projection, and its quantum version (M4) replaces KL with the Umegaki relative entropy you will meet in `02/09`.

## 2. Watching the asymmetry

Fix $p$ and slide a second distribution $q_t$ from some $q_0$ toward $p$. Both directions of KL must vanish when $q_t = p$ — but they approach zero along *different* curves.

```python
q0 = np.array([0.1, 0.2, 0.7])
ts = np.linspace(0.0, 1.0, 100)
forward = [kl_divergence(p, (1 - t) * q0 + t * p) for t in ts]   # D(p || q_t)
backward = [kl_divergence((1 - t) * q0 + t * p, p) for t in ts]  # D(q_t || p)

fig, ax = plt.subplots(figsize=(8, 4.5))
ax.plot(ts, forward, color=viz.SOURCE_COLOR, lw=2, label="D(p || q_t)")
ax.plot(ts, backward, color=viz.TARGET_COLOR, lw=2, label="D(q_t || p)")
ax.set_xlabel("t   (q_t slides from q_0 at t=0 to p at t=1)")
ax.set_ylabel("KL divergence (bits)")
ax.set_title("KL is asymmetric and vanishes only when the distributions match", pad=12)
ax.legend()
plt.show()
```

**Read the figure.** Both curves fall to exactly $0$ at $t = 1$, where $q_t = p$ — KL is zero only at a perfect match. But the two curves are **different**: the price of mistaking $p$ for $q$ is not the price of mistaking $q$ for $p$. That asymmetry is a feature, not a flaw — it is why we will sometimes need to choose *which way round* to write a KL projection.
