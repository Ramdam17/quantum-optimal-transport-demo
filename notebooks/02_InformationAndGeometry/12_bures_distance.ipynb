## 1. Bures distance — the quantum lift of Fisher–Rao

The Fisher–Rao distance of `02/07` becomes the **Bures distance** in the quantum setting:

$$ d_B(\rho, \sigma) = \sqrt{2\,(1 - F_U(\rho, \sigma))}, \qquad F_U = \operatorname{tr}\sqrt{\sqrt{\rho}\,\sigma\sqrt{\rho}}\ \ (\text{Uhlmann fidelity}). $$

For *diagonal* states $\rho = \operatorname{diag}(p)$, $\sigma = \operatorname{diag}(q)$, the Uhlmann fidelity is $F_U = \sum_i \sqrt{p_i q_i}$ — the **Bhattacharyya coefficient** of `02/07` — and Bures becomes the Hellinger distance, the small-angle cousin of Fisher–Rao. For *non-commuting* states it reaches beyond what the simplex can see.

```python
pairs = {
    "|+> vs |+>": (density_matrix(KET_PLUS), density_matrix(KET_PLUS)),
    "|0> vs |1>": (density_matrix(KET_0), density_matrix(KET_1)),
    "|+> vs |->": (density_matrix(KET_PLUS), density_matrix(KET_MINUS)),
    "|+> vs I/2": (density_matrix(KET_PLUS), maximally_mixed(2)),
    "|0> vs I/2": (density_matrix(KET_0), maximally_mixed(2)),
}
dists = {k: bures_distance(*v) for k, v in pairs.items()}
for k, d in dists.items():
    print(f"d_B({k:<12s}) = {d:.4f}")

names = list(dists)
fig, ax = plt.subplots(figsize=(8.5, 4.5))
ax.bar(names, [dists[n] for n in names], color=COLORS["quantum"], alpha=0.9)
ax.axhline(np.sqrt(2.0), color=COLORS["highlight"], linestyle="--", label="max = sqrt(2) (orthogonal)")
ax.set_ylabel("Bures distance")
ax.set_title("Orthogonal states reach sqrt(2); |+> vs I/2 is > 0 despite equal Z-diagonals", pad=12)
ax.legend()
plt.show()
```

**Read the figure and output.** Identical states sit at $0$; orthogonal pure states ($|0\rangle$ vs $|1\rangle$, $|+\rangle$ vs $|-\rangle$) reach the maximum $\sqrt{2} \approx 1.414$. And $|+\rangle$ vs $I/2$ — which share the *same* $Z$-diagonal — is strictly positive: Bures **sees the coherence**, exactly as the Umegaki relative entropy did in `02/09`. The infinitesimal version of $d_B^2$ is the *quantum Fisher information*, the metric that sets the ultimate precision of quantum estimation.
