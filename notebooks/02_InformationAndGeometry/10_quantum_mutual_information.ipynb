# 10 — Quantum Mutual Information

Lift mutual information to density matrices — and find a Bell pair shares twice what any classical pair can.

**Prerequisites:** `01/10_entanglement`, `02/03_mutual_information`, `02/09_umegaki_relative_entropy`.

**What you'll be able to do**
- Compute the quantum mutual information $I(A{:}B) = S(\rho_A) + S(\rho_B) - S(\rho_{AB})$.
- Show it is $0$ for a product state and $2$ bits for a Bell pair.
- Explain why $2$ bits is impossible classically (the classical bound is $1$ bit for two bits).

Classically, mutual information measures shared structure (`02/03`) and can never exceed the entropy of either variable. Quantum mechanics breaks that ceiling: for an entangled pair the whole is *more* correlated than its parts could classically allow — and the number is exactly double.

```python
import numpy as np
import matplotlib.pyplot as plt

from qot_course import viz
from qot_course.colors import COLORS
from qot_course.infotheory.quantum import quantum_mutual_information
from qot_course.quantum.composite import bell_state, tensor
from qot_course.quantum.density import density_matrix
from qot_course.quantum.states import KET_0, KET_PLUS

np.random.seed(42)
viz.use_course_style()
```

## 1. Quantum mutual information

By analogy with $I(X;Y) = D(p_{XY} \| p_X p_Y)$, the **quantum mutual information** is

$$ I(A{:}B) = S(\rho_A) + S(\rho_B) - S(\rho_{AB}). $$

For a product state it is $0$. For a Bell state, watch closely: $\rho_{AB}$ is *pure* (entropy $0$), yet each reduced state is maximally mixed (entropy $1$ each). So $I(A{:}B) = 1 + 1 - 0 = \mathbf{2}$ bits — double the classical bound, since classically $I(X;Y) \le \min(H(X), H(Y)) \le 1$ bit for two bits.

```python
rho_bell = density_matrix(bell_state())
rho_product = density_matrix(tensor(KET_0, KET_PLUS))

i_bell = quantum_mutual_information(rho_bell, dims=[2, 2])
i_prod = quantum_mutual_information(rho_product, dims=[2, 2])
print(f"I(A:B)  Bell state    = {i_bell:.3f} bit")
print(f"I(A:B)  product state = {i_prod:.3f} bit")

fig, ax = plt.subplots(figsize=(6.5, 4.5))
ax.bar(["product", "Bell"], [i_prod, i_bell], color=[COLORS["muted"], COLORS["quantum"]], alpha=0.9)
ax.axhline(1.0, color=COLORS["highlight"], linestyle="--", label="classical maximum (1 bit)")
ax.set_ylabel("I(A:B)  (bits)")
ax.set_title("A Bell pair doubles the classical mutual-information bound", pad=12)
ax.legend()
plt.show()
```

**Read the figure.** The product state shares $0$ bits. The Bell pair shares $2$ bits — clearing the classical maximum (dashed) by a factor of two — while each qubit alone is maximally uncertain. The whole knows what the parts cannot. That excess is a direct signature of entanglement, with no classical counterpart.
