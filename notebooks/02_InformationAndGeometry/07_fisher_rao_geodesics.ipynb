The Fisher metric told us the simplex is curved. A curved space has curved shortest paths — geodesics. The beautiful fact: under the change of variable $\phi_i = \sqrt{p_i}$, the Fisher–Rao geometry becomes the *round sphere*, so geodesics are simply great-circle arcs. We will compute distances and draw the arcs, and watch them bow away from the boundary exactly as the metric predicted.

```python
import numpy as np
import matplotlib.pyplot as plt

from qot_course import viz
from qot_course.geometry.info_geometry import (
    fisher_rao_distance, fisher_rao_geodesic, mixture_interpolation,
)

np.random.seed(42)
viz.use_course_style()
```

## 1. The Fisher–Rao distance

The Fisher–Rao geometry on the simplex has a closed form. The change of variable $\phi_i = \sqrt{p_i}$ maps the simplex onto a piece of the unit sphere $\{\phi : \sum_i \phi_i^2 = 1\}$, and the Fisher metric becomes the ordinary round-sphere metric. The geodesic distance is then the **angle** between sqrt-vectors (Rao 1945; Bhattacharyya 1943):

$$ d_{FR}(p, q) = 2\arccos\!\Big(\sum_i \sqrt{p_i\, q_i}\Big) \in [0, \pi]. $$

The quantity $\sum_i \sqrt{p_i q_i}$ is the **Bhattacharyya coefficient**. Two distributions reach the maximum distance $\pi$ exactly when their supports are disjoint.

```python
uniform = np.array([1 / 3, 1 / 3, 1 / 3])
print(f"d_FR(uniform, [0.7, 0.2, 0.1]) = {fisher_rao_distance(uniform, [0.7, 0.2, 0.1]):.3f}")
print(f"d_FR([1, 0, 0], [0, 1, 0])     = {fisher_rao_distance([1, 0, 0], [0, 1, 0]):.3f}   (= pi, disjoint support)")
print(f"d_FR(p, p)                     = {fisher_rao_distance([0.5, 0.3, 0.2], [0.5, 0.3, 0.2]):.3f}")
```

**Read the output.** Identical distributions are at distance $0$; distributions with **disjoint supports** sit at the maximum $\pi$ (their sqrt-vectors are orthogonal); everything else lies in between. Now let's *draw* a geodesic.

## 2. Drawing the geodesics

Geodesics are great-circle arcs (slerp) on the sphere of $\sqrt{p}$, projected back by squaring. Let's draw a Fisher–Rao geodesic between two points near opposite edges and compare it to the straight **mixture** path $(1-t)p + tq$.

```python
p0 = np.array([0.10, 0.85, 0.05])
p1 = np.array([0.85, 0.10, 0.05])
p2 = np.array([0.05, 0.05, 0.90])

ts = np.linspace(0.0, 1.0, 80)
fr_arc = np.array([fisher_rao_geodesic(p0, p1, t) for t in ts])
linear_path = np.array([mixture_interpolation(p0, p1, t) for t in ts])
fr_arc_to_top = np.array([fisher_rao_geodesic(p0, p2, t) for t in ts])

viz.plot_simplex_paths(
    {
        "Fisher-Rao geodesic ($p_0 \\to p_1$)": fr_arc,
        "linear / mixture path ($p_0 \\to p_1$)": linear_path,
        "Fisher-Rao geodesic ($p_0 \\to p_2$)": fr_arc_to_top,
    },
    endpoints={"$p_0$": p0, "$p_1$": p1, "$p_2$": p2},
    title="Fisher-Rao geodesics vs the linear (mixture) path on the 2-simplex",
)
plt.show()
```

**Read the figure.** Compare the Fisher–Rao arc to the straight mixture line between $p_0$ and $p_1$: the geodesic **bows away from the boundary** (here, away from the $\delta_1$–$\delta_2$ edge). That bow is exactly the curvature we read off the Bernoulli plot in `02/06` — the metric stretches near the edge, so the shortest path cuts through the interior instead. The mixture path treats the simplex as flat; the Fisher–Rao geodesic treats it as curved.
