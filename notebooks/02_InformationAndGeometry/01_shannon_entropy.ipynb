```python
import numpy as np
import matplotlib.pyplot as plt

from qot_course import viz
from qot_course.infotheory.classical import shannon_entropy

np.random.seed(42)
viz.use_course_style()
```

## 1. Surprise and entropy

A rare event is surprising; a certain event is not. Shannon made this precise: the surprise of an outcome of probability $p$ is $-\log_2 p$ (in bits), and the **entropy** is the *average* surprise,

$$ H(p) = -\sum_x p(x)\log_2 p(x). $$

A fair coin carries $1$ bit; a biased coin carries less, because its outcomes are more predictable. Let's watch entropy change as we bias a coin from certain to fair to certain again.

```python
bias = np.linspace(0.001, 0.999, 200)
entropy_curve = [shannon_entropy([b, 1 - b]) for b in bias]

fig, ax = plt.subplots(figsize=(8, 4.5))
ax.plot(bias, entropy_curve, color=viz.SOURCE_COLOR)
ax.axvline(0.5, color=viz.TARGET_COLOR, linestyle="--", label="fair coin (1 bit)")
ax.set_xlabel("coin bias  P(heads)")
ax.set_ylabel("entropy  H  (bits)")
ax.set_title("Entropy is maximal for a fair coin", pad=12)
ax.legend()
plt.show()
```

**Read the figure.** Entropy peaks at $1$ bit for the fair coin (maximum uncertainty) and falls to $0$ at the certain outcomes $p=0$ and $p=1$ (no surprise at all). Uncertainty *is* information: the more unpredictable a source, the more bits you need to describe its output. This one curve is the seed of data compression and channel capacity — and, for us, of the entropic regularisation that later makes optimal transport fast.
