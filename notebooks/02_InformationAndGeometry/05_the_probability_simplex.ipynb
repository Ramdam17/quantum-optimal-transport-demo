We have been computing *with* distributions. Now we step back and look at the *space* of all of them. For three outcomes that space is a triangle — the 2-simplex — and every distribution is a single point inside it. The rest of this arc asks a surprisingly deep question about that triangle: how far apart are two distributions? The answer is not what flat, Euclidean intuition suggests.

```python
import numpy as np
import matplotlib.pyplot as plt

from qot_course import viz

np.random.seed(42)
viz.use_course_style()
```

## 1. The 2-simplex

A categorical distribution over $K$ outcomes is a vector $p \in \mathbb{R}^K$ with $p_i \ge 0$ and $\sum_i p_i = 1$. For $K = 3$ those constraints carve out a *triangle* — the **2-simplex**:

- the three **corners** $\delta_1, \delta_2, \delta_3$ are the certain outcomes (all mass on one symbol);
- the **centre** $(1/3, 1/3, 1/3)$ is the uniform distribution — maximum entropy, from `02/01`;
- the **edges** are the slices where one outcome is forbidden (one mass is zero).

Every point inside the triangle *is* a distribution. Let's place a few.

```python
examples = {
    "uniform $(1/3, 1/3, 1/3)$": np.array([1 / 3, 1 / 3, 1 / 3]),
    "skewed $(0.7, 0.2, 0.1)$": np.array([0.7, 0.2, 0.1]),
    "almost $\\delta_1$": np.array([0.92, 0.04, 0.04]),
    "on an edge $(0.5, 0.5, 0)$": np.array([0.5, 0.5, 0.0]),
}
viz.plot_simplex_points(examples)
plt.show()
```

**Read the figure.** Each point is a distribution. The vertices are the deterministic outcomes; the centre is uniform; the edges are the boundary where one outcome has been ruled out. The "almost $\delta_1$" point sits near a corner — barely uncertain. Our question for the rest of the arc: **what is the natural way to measure distance in this triangle?** There is more than one answer, and the difference will carry us all the way to quantum optimal transport.
